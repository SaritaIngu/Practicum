```python
import pandas as pd
import numpy as np
```

Start with dataframe with ticker, date, title, link and text. Code for reading in file can be removed once connected to article text data directly.

```python
def Sentiment(a):
    import pysentiment as ps
    lm = ps.LM()
    tokens = lm.tokenize(a)
    a= lm.get_score(tokens)['Polarity']
    if a>0:
        return "Positive" 
    elif(a==0):
        return "Neutral" 
    else:
        return "Negative" 
    
def Score(a):
    import pysentiment as ps
    lm = ps.LM()
    tokens = lm.tokenize(a)
    a= lm.get_score(tokens)['Polarity']
    return a
```
