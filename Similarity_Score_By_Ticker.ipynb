### Similarity Scorer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
```

Start with dataframe with ticker, date, title, link and text. Code for reading in file can be removed once connected to article text data directly.

```python
df = pd.read_csv("WSJoutput5.csv",header = None, encoding = "ISO-8859-1")
```

```python
df.columns = ['Ticker','Date','Title','Link','Text']
```

Set "articles" equal to column with text data.

```python
articles = df['Text'].dropna()
#articles
```

```python
vectorizer = TfidfVectorizer(min_df = 1, stop_words = 'english', use_idf=False)
dtm = vectorizer.fit_transform(articles)
doc_term_matrix = pd.DataFrame(dtm.toarray(),index=articles,columns=vectorizer.get_feature_names())
```

```python
similarity = np.asarray(np.asmatrix(doc_term_matrix) * np.asmatrix(doc_term_matrix).T)
similarity = np.triu(similarity, k=1)
df_sim = pd.DataFrame(similarity, columns = articles)
```

Set threshold for what is considered similar. Anything above this threshold will be dropped.

```python
similarity_threshold = 0.65
```

```python
unique_articles = df_sim[(df_sim<=similarity_threshold)].dropna(axis = 1, how = 'any').columns.values
```

```python
df['Unique'] = df['Text'].apply(lambda x: 1 if x in unique_articles else 0)
```

```python
df_unique = df[df['Unique']==1]
```
